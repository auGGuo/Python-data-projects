```python
import pandas as pd
```

```python
url = 'https://raw.githubusercontent.com/justmarkham/pandas-videos/master/data/ufo.csv'
ufo = pd.read_csv(url)
ufo.head()
```

```python
ufo.columns
```

```python
#Method 1: Renaming a single column
# inplace=True to affect DataFrame

ufo.rename(columns={'Colors Reported':'Colors_Reported'}, inplace=True)
ufo.head()
    
```

```python
#Method 2: Renaming multiple columns

ufo.columns = {'city', 'colors reported', 'shape reported', 'state', 'time'}
ufo.head()
    
```

```python
#Method 3: Change columns while importing the data file

ufo = pd.read_csv(url, names=['City', 'Colors_Reported', 'Shape_Reported', 'State', 'Time'], header=0)
ufo.head()
```

```python
#Method 4: Replacing spaces with underscores for all columns

ufo.columns = ufo.columns.str.replace('_', ' ')
ufo.head()
```
