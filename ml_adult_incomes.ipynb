```python
X['native.country'] = ['United-States' if x == 'United-States' else 'Others' for x in X['native.country']]
X['native.country'].value_counts()
```

```python
todummy_list = []
for col in X.columns:
     if X[col].dtype == 'O':
            todummy_list.append(col)
            
todummy_list
```

```python
pd.get_dummies(X['education'], prefix='education', dummy_na=False).head()
```

```python
def dummy_df(df, todummy_list):
    for x in todummy_list:
        dummies = pd.get_dummies(X[x], prefix=x, dummy_na=False)
        df = df.drop(x, axis=1)
        df = pd.concat([df, dummies], axis=1)
    return df
```

```python
X = dummy_df(X, todummy_list)
```
