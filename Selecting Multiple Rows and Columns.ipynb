```python
import pandas as pd
```

```python
url = 'https://raw.githubusercontent.com/justmarkham/pandas-videos/master/data/ufo.csv'

ufo = pd.read_csv(url)
ufo.head()
```

```python
# show first 10 shows

ufo.head(10)
```

```python
# .loc DataFrame method
# filtering rows and selecting columns by label

# format
# ufo.loc[rows, columns]

# row 0, all columns

ufo.loc[0, :]
```

```python
# rows 0,  2
# all columns

ufo.loc[[0, 2], :]
```

```python
# rows 0, 1, 2
# all columns

ufo.loc[0:2, :]
```

```python
# all rows
# column: City
ufo.loc[:, 'City']
```

```python
# all rows
# column: City, State

ufo.loc[:, ['City', 'State']]
```

```python
# multiple rows and multiple columns

ufo.loc[0:2, 'City':'State']
```

```python
# filter using City=='Oakland'

ufo[ufo['City'] == 'Oakland']
```

```python
# easier-to-read code
# here you specify the rows and columns you want
# ufo.loc[rows, columns]

ufo.loc[ufo['City'] == 'Oakland', :]
```

```python
# again, specifying the rows and columns you want
# this would be the best way to do it compared to chain indexing 

ufo.loc[ufo.City=='Oakland', 'State']
```
