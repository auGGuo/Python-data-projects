### Step1 unpack the ratings data

```python
# find out what rating data looks like
ted.loc[0, 'ratings']
```

```python
import ast
```

```python
ast.literal_eval(ted.loc[0, 'ratings'])[0]
```

```python
def str_to_list(ratings_str):
    return ast.literal_eval(ratings_str)

ted.ratings.apply(str_to_list).head()
```

```python
# use lambda method

ted['ratings_list'] = ted.ratings.apply(lambda x: ast.literal_eval(x))
ted.head()
```

### Step2 count the total of ratings received by each talk

```python
# slow way
ted['rating_count'] = 0
ted.head()

for index in range(ted.shape[0]):
    for list_index in range(len(ted.ratings_list[index])):
        ted.loc[index, 'rating_count'] += ted.ratings_list[index][list_index]['count']
        
ted.head()
```

```python
# better way

def get_number_rating(list_of_dict):
    num = 0
    for dic in list_of_dict:
        num += dic['count']
    return num
```

```python
get_number_rating(ted.loc[0, 'ratings_list'])
```

```python
ted['rating_count2'] = ted.ratings_list.apply(get_number_rating)
ted.head()
```

```python
# double_check to see if there is any mistakes
ted.rating_count.describe()
```

### Step3 which ocupation delivers the most funniest Ted talks on average

```python
# first find out the funny rating count

def get_funny_rating(list_of_dict):
    for dic in list_of_dict:
        if dic['name'] == 'Funny':
            return dic['count']

ted['funny_rating_count'] = ted.ratings_list.apply(get_funny_rating)
ted.head()       
```

```python
# find the the percentage of ratings are funny 

ted['funny_rate'] = ted.funny_rating_count / ted.rating_count
ted.head()
```

### Step4 analyze funny rate by occupation 

```python
ted.groupby('speaker_occupation').funny_rate.mean().sort_values().tail()
```

```python
ted.speaker_occupation.describe()
```

### Step5 focus on occupations that are well represented in the data 

```python
ted.speaker_occupation.value_counts()
```

```python
# filter out occupation count that are less than 5 
occupation_count = ted.speaker_occupation.value_counts()
occupation_count[occupation_count >= 5]
```

```python
top_occupations = occupation_count[occupation_count >= 5].index
top_occupations
```

```python
# output the dataframe only related to top occupations
ted_top_occupations = ted[ted.speaker_occupation.isin(top_occupations)]
ted_top_occupations 
```

```python
ted_top_occupations.groupby('speaker_occupation').mean().funny_rate.sort_values()
```
