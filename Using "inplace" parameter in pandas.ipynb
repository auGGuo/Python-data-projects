```python
import pandas as pd
```

```python
url = 'https://raw.githubusercontent.com/justmarkham/pandas-videos/master/data/ufo.csv'

ufo = pd.read_csv(url)
```

```python
ufo.shape
```

```python
ufo.head()
```

```python
# dropping City column

ufo.drop('City', axis=1).head()
```

```python
# you can see that the City column is not gone 
# drop() method has inplace=False as default

ufo.head()
```

```python
# you want to change to inplace=True to affect the underlying data

ufo.drop('City', axis=1, inplace=True)
```

```python
ufo.head()
```

```python
# dropna with how='any' would drop any row with 'NaN'

ufo.dropna(how='any')
```

```python
ufo.shape
```

```python
# when you can not use inplace=True, you can use an assignment instead

ufo = ufo.set_index('Time')
```

```python
ufo.head()
```

```python
ufo.tail()
```

```python
ufo.fillna(method='ffill').head()
```
