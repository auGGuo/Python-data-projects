```python
path = 'UN_All_CountryFiles/Canada.xlsx'
df_can = pd.read_excel(path, sheet_name='Canada by Citizenship', skiprows=20, skipfooter=2)
df_can.head()
```

```python
drop_cols = ['Type', 'Coverage', 'AREA', 'REG', 'DEV']
df_can.drop(columns=drop_cols, axis=1, inplace=True )
df_can.head()
```

```python
df_can.rename(columns={'OdName':'Country', 'AreaName':'Continent', 'RegName':'Region'}, inplace=True)
df_can.head()
```

```python
df_can['Total'] = df_can.sum(axis=1, skipna=True)
df_can.head()
```

```python
df_can.columns[4:13].astype('float')
df_can.columns.values
```

```python
df_can.columns = list(map(str, df_can.columns))
print([type(x) for x in df_can.columns])
```

```python
df_can.set_index('Country', inplace=True)
```

```python
df_can.index.name = None
df_can.head()
```

```python
df_can[df_can.index == 'Japan'].T.squeeze()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.style.use('ggplot')
```

```python
years = list(map(str, range(1980, 2014)))
years
```

```python
# find out the immigration info about haiti

haiti = df_can.loc['Haiti', years]
```

```python
haiti.plot(kind = 'line')
```

```python
haiti.index = haiti.index.map(int)
haiti.index
```

```python
# We can clearly notice how number of immigrants from Haiti spiked
# up from 2010 as Canada stepped up its efforts to accept refugees from Haiti.
# annotate the 2010 Earthquake. 

haiti.plot(kind='line')
plt.title('Immigration from Haiti')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')

plt.text(1998, 6000, '2010 Haiti Earthquake')

plt.show()
```

Question: Let's compare the number of immigrants from India and China from 1980 to 2013.

```python
df_CI = df_can.loc[['China', 'India'], years]
df_CI
```

```python
df_CI.plot(kind='line')
```

The above graph doesn't look right. Matplotlib plots indices var on the x axis and column var on the y axis.Therefore, the dataframe needs to be transposed. 

```python
df_CI = df_CI.transpose()
df_CI
```

```python
df_CI.plot(kind='line')
plt.xlabel('Year')

plt.show()
```
