```python
import pandas as pd
```

```python
link = 'https://raw.githubusercontent.com/justmarkham/pandas-videos/master/data/ufo.csv'

ufo = pd.read_csv(link)
ufo.head()
```

```python
# to get 3 random rows
# each time you run this, you would have 3 different rows

ufo.sample(n=3)
```

```python
ufo.sample(n=3)
```

```python
# you can use random_state for reproducibility

ufo.sample(n=3, random_state=2)
```

```python
ufo.sample(n=3, random_state=2)
```

```python
# fraction of rows
# here you get 10% of the rows

ufo.sample(frac=0.1, random_state=3)
```

```python
ufo.sample(frac=0.1, random_state=3)
```
