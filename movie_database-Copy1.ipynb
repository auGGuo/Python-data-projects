```python
my_votes = {
    "Star Wars": 9,
    "Paris is Burning": 8,
    "Dead Poets Society": 7,
    "The Empire Strikes Back": 9.5,
    "The Shining": 8,
    "Return of the Jedi": 8,
    "1941": 8,
    "Forrest Gump": 7.5,
}
```

```python
vote_comparison = pd.Series(my_votes).to_frame()
vote_comparison.columns = ['my_rating']
vote_comparison
```

```python
movie_list = vote_comparison.index.values.tolist()
movie_list
```

```python
movies.set_index('title', inplace=True)
```

```python
movies.head()
```

```python
vote_avg = movies.loc[movie_list].vote_average
vote_avg
```

```python
vote_comparison = pd.concat([vote_comparison, vote_avg], axis=1)
vote_comparison
```

```python
vote_comparison.index.name = None
```

```python
vote_comparison['percent_diff'] = (vote_comparison['vote_average'] - vote_comparison['my_rating']) / vote_comparison['my_rating'] * 100
```
