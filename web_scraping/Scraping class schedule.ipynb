```python
def get_classes_from_website():
    classes = pd.DataFrame()
    
    for num in range(1,3):
        website = 'https://ischool.syr.edu/classes/?page=' + str(num)
        data_list = pd.read_html(website)
        df = data_list[0]
        classes = pd.concat([df, classes])
    return classes 

classes = get_classes_from_website()       
classes.head()
```

```python
# create a new column identify grad and undergrad course 
# grad courses are IST5XX or above

def get_course_level(course_id):
    num = course_id[3:]
    num = int(num)
    if num >= 500:
        return 'Grad'
    else:
        return 'Undergrad'

classes['Level'] = classes.Course.apply(get_course_level)
classes.sample(10)
```

```python
# want to take grad classes
# don't want classes on Sunday

grad_classes = classes[classes.Level == 'Grad']
grad_classes_no_sunday = grad_classes[grad_classes.Day.str.contains(r'Su', na=False)]
grad_classes_no_sunday
```
