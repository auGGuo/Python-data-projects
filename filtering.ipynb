```python
import pandas as pd
url = 'https://raw.githubusercontent.com/justmarkham/pandas-videos/master/data/imdb_1000.csv'
movies = pd.read_csv(url)
movies.head()
```

```python
movies.shape
```

```python
booleans = []

for length in movies['duration']:
    if length >= 200:
        booleans.append(True)
    else:
        booleans.append(False)

booleans[:5]
```

```python
#convert booleans into pandas serises

is_long = pd.Series(booleans)

is_long.head()
```

```python
# this pulls out duration >= 200mins

movies[is_long]
```

```python
#Faster method without a for loop

isVeryLong = movies['duration'] >= 220
movies[isVeryLong] 
```

```python
# Even better way to simplify movies[is_long]

movies[movies.duration > 230]
```

```python
# if we want to study the duration and only the genre instead of all the columns

movies[movies['duration'] > 200].genre
```

```python
# better practice 

movies.loc[movies.duration > 200, 'genre']
```
